```python
import pandas as pd
import numpy as np
import glob, re
```

```python
weather = pd.read_csv('../input/air_with_weather.csv', index_col = 0)
```

```python
weather = weather[['air_store_id', 'visit_date', 'avg_temperature', 'high_temperature', 'low_temperature',
       'precipitation', 'avg_wind_speed']]
```

```python
weather.describe()
```

```python
def f(x):
    if x['low_temperature'] < 0 or x['high_temperature'] > 30:
        return 1
    else:
        return 0
```

```python
weather["extreme_temp"] = weather.apply(f, axis=1)
weather['rainy'] = weather['precipitation'] > 50
weather['windy'] = weather['avg_wind_speed'] > 5
weather['rainy'] = weather['rainy'].astype(int)
weather['windy'] = weather['windy'].astype(int)
```

```python
weather
```

```python
weather.to_csv("weather.csv", index=False)
```
