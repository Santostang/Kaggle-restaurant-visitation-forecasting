```python
import pandas as pd
import numpy as np
```

## Update air visits

```python
# read all the data
air_visits = pd.read_csv('../input/air_visit_data.csv')
```

```python
# transfer air_visits
air_visits['visit_date'] = pd.to_datetime(air_visits['visit_date'])
air_visits['dow'] = air_visits['visit_date'].dt.dayofweek
air_visits['year'] = air_visits['visit_date'].dt.year
air_visits['month'] = air_visits['visit_date'].dt.month
air_visits['visit_date'] = air_visits['visit_date'].dt.date
```

```python
air_visits['lastmonth'] = air_visits['visit_date'] - pd.DateOffset(months=1)
air_visits['nextmonth'] = air_visits['visit_date'] + pd.DateOffset(months=1)
air_visits['next2month'] = air_visits['visit_date'] + pd.DateOffset(months=1)
```

```python
air_visits['year_month'] = air_visits['visit_date'].map(lambda x: x.strftime('%Y%m'))
air_visits['last_year_month'] = air_visits['lastmonth'].map(lambda x: x.strftime('%Y%m'))
air_visits['next_year_month'] = air_visits['nextmonth'].map(lambda x: x.strftime('%Y%m'))
air_visits['next2_year_month'] = air_visits['next2month'].map(lambda x: x.strftime('%Y%m'))
```

```python
tmp = air_visits.groupby(['air_store_id','dow', 'year_month'], as_index=False)['visitors'].min().rename(columns={'visitors':'lastm_min_visitors', 'year_month':'new_year_month'})
air_visits = pd.merge(air_visits, tmp, how = 'left', left_on = ['air_store_id', 'dow', 'last_year_month'], right_on = ['air_store_id', 'dow', 'new_year_month'])
air_visits.drop(['new_year_month'], inplace=True, axis=1)
tmp = air_visits.groupby(['air_store_id','dow', 'year_month'], as_index=False)['visitors'].mean().rename(columns={'visitors':'lastm_mean_visitors', 'year_month':'new_year_month'})
air_visits = pd.merge(air_visits, tmp, how = 'left', left_on = ['air_store_id', 'dow', 'last_year_month'], right_on = ['air_store_id', 'dow', 'new_year_month'])
air_visits.drop(['new_year_month'], inplace=True, axis=1)
tmp = air_visits.groupby(['air_store_id','dow', 'year_month'], as_index=False)['visitors'].median().rename(columns={'visitors':'lastm_median_visitors', 'year_month':'new_year_month'})
air_visits = pd.merge(air_visits, tmp, how = 'left', left_on = ['air_store_id', 'dow', 'last_year_month'], right_on = ['air_store_id', 'dow', 'new_year_month'])
air_visits.drop(['new_year_month'], inplace=True, axis=1)
tmp = air_visits.groupby(['air_store_id','dow', 'year_month'], as_index=False)['visitors'].max().rename(columns={'visitors':'lastm_max_visitors', 'year_month':'new_year_month'})
air_visits = pd.merge(air_visits, tmp, how = 'left', left_on = ['air_store_id', 'dow', 'last_year_month'], right_on = ['air_store_id', 'dow', 'new_year_month'])
air_visits.drop(['new_year_month'], inplace=True, axis=1)
```

```python
this_month = air_visits[['air_store_id','dow', 'year_month', 'visitors']]
next_month = air_visits[['air_store_id','dow', 'next_year_month', 'visitors']].rename(columns={ 'next_year_month':'year_month'})
combine_two_month = pd.concat([this_month, next_month], ignore_index=True)
```

```python
tmp = combine_two_month.groupby(['air_store_id','dow', 'year_month'], as_index=False)['visitors'].min().rename(columns={'visitors':'last_twom_min_visitors', 'year_month':'new_year_month'})
air_visits = pd.merge(air_visits, tmp, how = 'left', left_on = ['air_store_id', 'dow', 'last_year_month'], right_on = ['air_store_id', 'dow', 'new_year_month'])
air_visits.drop(['new_year_month'], inplace=True, axis=1)
tmp = combine_two_month.groupby(['air_store_id','dow', 'year_month'], as_index=False)['visitors'].mean().rename(columns={'visitors':'last_twom_mean_visitors', 'year_month':'new_year_month'})
air_visits = pd.merge(air_visits, tmp, how = 'left', left_on = ['air_store_id', 'dow', 'last_year_month'], right_on = ['air_store_id', 'dow', 'new_year_month'])
air_visits.drop(['new_year_month'], inplace=True, axis=1)
tmp = combine_two_month.groupby(['air_store_id','dow', 'year_month'], as_index=False)['visitors'].median().rename(columns={'visitors':'last_twom_median_visitors', 'year_month':'new_year_month'})
air_visits = pd.merge(air_visits, tmp, how = 'left', left_on = ['air_store_id', 'dow', 'last_year_month'], right_on = ['air_store_id', 'dow', 'new_year_month'])
air_visits.drop(['new_year_month'], inplace=True, axis=1)
tmp = combine_two_month.groupby(['air_store_id','dow', 'year_month'], as_index=False)['visitors'].max().rename(columns={'visitors':'last_twom_max_visitors', 'year_month':'new_year_month'})
air_visits = pd.merge(air_visits, tmp, how = 'left', left_on = ['air_store_id', 'dow', 'last_year_month'], right_on = ['air_store_id', 'dow', 'new_year_month'])
air_visits.drop(['new_year_month'], inplace=True, axis=1)
```

```python
this_month = air_visits[['air_store_id','dow', 'year_month', 'visitors']]
next_month = air_visits[['air_store_id','dow', 'next_year_month', 'visitors']].rename(columns={ 'next_year_month':'year_month'})
next2_month = air_visits[['air_store_id','dow', 'next2_year_month', 'visitors']].rename(columns={ 'next2_year_month':'year_month'})
combine_three_month = pd.concat([this_month, next_month, next2_month], ignore_index=True)
```

```python
tmp = combine_three_month.groupby(['air_store_id','dow', 'year_month'], as_index=False)['visitors'].min().rename(columns={'visitors':'last_threem_min_visitors', 'year_month':'new_year_month'})
air_visits = pd.merge(air_visits, tmp, how = 'left', left_on = ['air_store_id', 'dow', 'last_year_month'], right_on = ['air_store_id', 'dow', 'new_year_month'])
air_visits.drop(['new_year_month'], inplace=True, axis=1)
tmp = combine_three_month.groupby(['air_store_id','dow', 'year_month'], as_index=False)['visitors'].mean().rename(columns={'visitors':'last_threem_mean_visitors', 'year_month':'new_year_month'})
air_visits = pd.merge(air_visits, tmp, how = 'left', left_on = ['air_store_id', 'dow', 'last_year_month'], right_on = ['air_store_id', 'dow', 'new_year_month'])
air_visits.drop(['new_year_month'], inplace=True, axis=1)
tmp = combine_three_month.groupby(['air_store_id','dow', 'year_month'], as_index=False)['visitors'].median().rename(columns={'visitors':'last_threem_median_visitors', 'year_month':'new_year_month'})
air_visits = pd.merge(air_visits, tmp, how = 'left', left_on = ['air_store_id', 'dow', 'last_year_month'], right_on = ['air_store_id', 'dow', 'new_year_month'])
air_visits.drop(['new_year_month'], inplace=True, axis=1)
tmp = combine_three_month.groupby(['air_store_id','dow', 'year_month'], as_index=False)['visitors'].max().rename(columns={'visitors':'last_threem_max_visitors', 'year_month':'new_year_month'})
air_visits = pd.merge(air_visits, tmp, how = 'left', left_on = ['air_store_id', 'dow', 'last_year_month'], right_on = ['air_store_id', 'dow', 'new_year_month'])
air_visits.drop(['new_year_month'], inplace=True, axis=1)
```

```python
air_visits.columns
```

```python
air_visits.drop(['visitors', 'dow', 'year', 'month',
       'lastmonth', 'nextmonth', 'next2month', 'year_month', 'last_year_month',
       'next_year_month', 'next2_year_month'], inplace=True, axis=1)
```

```python
air_visits.to_csv('past_air_visit.csv', index=False)
```

```python
train = pd.merge(train, past_air_visit, how = 'left', on = ['air_store_id', 'visit_date'])
val = pd.merge(val, past_air_visit, how = 'left', on = ['air_store_id', 'visit_date'])
```
